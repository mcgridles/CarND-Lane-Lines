````python
#importing some useful packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline
````

````python
# Import everything needed to edit/save/watch video clips
import imageio
#imageio.plugins.ffmpeg.download()

from moviepy.editor import VideoFileClip
from IPython.display import HTML
````

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    return lines

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)

def slope(line):
    """
    Calculates the slope of a line
    """
    x1 = line[0]
    y1 = line[1]
    x2 = line[2]
    y2 = line[3]
    return (y2-y1)/(x2-x1)

def mean(data):
    """
    Calculates the mean of a list of data
    """
    try:
        return sum(data) / len(data)
    except ZeroDivisionError:
        # protects against images where all of the lines
        # where filtered out because the slopes were too small
        return 0
    
````

````python
class Processor():
    def __init__(self):
        self.prev_slope = 0
        
    def draw_lines(self, img, lines, color=[255, 0, 0], thickness=1):
        """
        Superimposes lane lines on an image
        """
        imshape = image.shape
        
        # separate lines
        left_line_slopes, left_line_x, left_line_y = [], [], []
        right_line_slopes, right_line_x, right_line_y = [], [], []
        for line in lines:
            line_slope = slope(line[0])
            if abs(line_slope) < 0.5 or abs(line_slope) > 4:
                # filter out very small or large slopes caused by
                # small lane lines and erroneous edges
                continue
            elif line_slope < 0:
                left_line_slopes.append(line_slope)
                left_line_x.append(line[0][0])
                left_line_x.append(line[0][2])
                left_line_y.append(line[0][1])
                left_line_y.append(line[0][3])
            else:
                right_line_slopes.append(line_slope)
                right_line_x.append(line[0][0])
                right_line_x.append(line[0][2])
                right_line_y.append(line[0][1])
                right_line_y.append(line[0][3])

        # average lists
        left_slope_avg = mean(left_line_slopes)
        left_x_avg = mean(left_line_x)
        left_y_avg = mean(left_line_y)
        right_slope_avg = mean(right_line_slopes)
        right_x_avg = mean(right_line_x)
        right_y_avg = mean(right_line_y)

        # extrapolate the lines to fille the region of interest
        left_lane_line = self.extrapolate(left_x_avg, left_y_avg, left_slope_avg, imshape)
        right_lane_line = self.extrapolate(right_x_avg, right_y_avg, right_slope_avg, imshape)
        lane_lines = [left_lane_line, right_lane_line]

        for line in lane_lines:
            for x1,y1,x2,y2 in line:
                cv2.line(img, (x1, y1), (x2, y2), color, thickness)
        
    def extrapolate(self, x_mean, y_mean, slope_mean, imshape):
        """
        Finds a line that fills up the region of interest based on a line
        found in the image
        """
        # if all the edges were filtered out, use the previous slope
        # until edges are found
        if slope_mean != 0:
            self.prev_slope = slope_mean
        else:
            slope_mean = self.prev_slope

        intercept = y_mean - slope_mean*x_mean

        # determine they y range for the lines
        y_min_endpoint = int(imshape[0]/2+50)
        y_max_endpoint = int(imshape[0])

        # calculate x at the limits of the y range
        x_min_endpoint = int((y_min_endpoint-intercept)/slope_mean)
        x_max_endpoint = int((y_max_endpoint-intercept)/slope_mean)

        return [[x_min_endpoint, y_min_endpoint, x_max_endpoint, y_max_endpoint]]

    def process_image(self, image):
        """
        Takes a color image, figures out where the lane lines are, and overlays
        a line on top of them to mark the lines
        """
        # get grayscaled image
        gray_img = grayscale(image)

        # apply Gaussian smoothing
        kernel_size = 3
        blur_gray_img = cv2.GaussianBlur(gray_img,(kernel_size, kernel_size),0)

        # run edge detection
        low_threshold = 50
        high_threshold = 150
        canny_img = canny(blur_gray_img, low_threshold, high_threshold)

        # determine region of interest
        imshape = image.shape
        lower_left = (0, imshape[0])
        upper_left = ((imshape[1]/2)-10, (imshape[0]/2)+50)
        upper_right = ((imshape[1]/2)+10, (imshape[0]/2)+50)
        lower_right = (imshape[1], imshape[0])
        vertices = np.array([[lower_left, upper_left, upper_right, lower_right]], dtype=np.int32)
        masked_img = region_of_interest(canny_img, vertices)

        # add hough lines
        rho = 1
        theta = (np.pi/180)
        threshold = 15
        min_line_length = 40
        max_line_gap = 30
        h_lines = hough_lines(masked_img, rho, theta, threshold, min_line_length, max_line_gap)
        
        # draw lines
        line_image = np.copy(image)*0
        self.draw_lines(line_image, h_lines, thickness=8)

        result = weighted_img(image, line_image) 
        return result
    
image = mpimg.imread('test_images/solidWhiteCurve.jpg')
#plt.imshow(image)
processor = Processor()
result = processor.process_image(image)
plt.imshow(result)
````
